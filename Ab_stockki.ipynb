```python
import yfinance as yf
import pandas as pd 
pd.options.mode.chained_assignment = None  # default='warn' #remove warning #i used this to remove warning

#Defining My ticker symbol 
ticker = 'AB'

#Collect ticker data using this 
tickerdata = yf.Ticker(ticker)

#1m means one month 
tickerDf = tickerdata.history(interval='1mo', start='2015-10-01', end='2020-12-01')

#lets check what we have got 

tickerDf.head(50)
```

```python
# lets drop some columns and Null values 
ty = tickerDf.dropna()

for col in ty.columns: 
    print(col) 
ty.drop(columns=['Open','High','Low','Volume','Dividends','Stock Splits'], inplace = True)

```

```python
#Lets take the percentage change or HDR of the stock
ty = ty.pct_change()
ty1 = ty.rename(columns={'Close': ticker })
ty1.dropna()
```

```python
cicker = '^GSPC'

#get data on this ticker
cickerdata = yf.Ticker(cicker)

#get the historical prices for this ticker
cickerDf = cickerdata.history(interval='1mo', start='2015-10-01', end='2020-12-01')

#see your data
cy = cickerDf.dropna()
cy.head
for col in cy.columns: 
    print(col) 
cy.drop(columns=['Open','High','Low','Volume','Dividends','Stock Splits'], inplace = True)
#for ind, row in cy.iterrows():
#    cy.loc[ind, "GSPCHDR"] =  cy.pct_change() 
cy = cy.pct_change()
cy1 = cy.rename(columns={'Close': cicker}) # or i could use '^gspc' for str
cy1.dropna()
```

```python
#lets get the maturity rates 
trs = pd.read_csv('https://raw.githubusercontent.com/asiqman12/ln_regression/main/w.csv')
trs
```

```python
#now lets combine
for ind, row in trs.iterrows():
    trs.loc[ind, "DGS1MO"] = row['DGS1MO'] / 100

import numpy as np
# pre 0.24
trs[ticker] = ty1[ticker].values
# >= 0.24
trs[ticker] = ty1[ticker].to_numpy()
trs[cicker] = cy1[cicker].values
# >= 0.24
trs[cicker] = cy1[cicker].to_numpy()
trs.dropna()

trs["Xaxis"] = trs[cicker] - trs['DGS1MO']
trs["Yaxis"] = trs[ticker] - trs['DGS1MO']
trs.dropna()
```
