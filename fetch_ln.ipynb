```python
import yfinance as yf
import pandas as pd 
pd.options.mode.chained_assignment = None  # default='warn' #remove warning

#define the ticker symbol
ticker = 'AB'

#get data on this ticker
tickerdata = yf.Ticker(ticker)

#get the historical prices for this ticker
tickerDf = tickerdata.history(interval='1mo', start='2015-10-01', end='2020-12-01')

#see your data

tickerDf.head(50)
ty = tickerDf.dropna()

for col in ty.columns: 
    print(col) 
ty.drop(columns=['Open','High','Low','Volume','Dividends','Stock Splits'], inplace = True)
ty = ty.pct_change()
ty1 = ty.rename(columns={'Close': 'firmcl'})
ty1.dropna()
cicker = '^GSPC'

#get data on this ticker
cickerdata = yf.Ticker(cicker)

#get the historical prices for this ticker
cickerDf = cickerdata.history(interval='1mo', start='2015-10-01', end='2020-12-01')

#see your data
cy = cickerDf.dropna()
cy.head
for col in cy.columns: 
    print(col) 
cy.drop(columns=['Open','High','Low','Volume','Dividends','Stock Splits'], inplace = True)
#for ind, row in cy.iterrows():
#    cy.loc[ind, "GSPCHDR"] =  cy.pct_change() 
cy = cy.pct_change()
cy1 = cy.rename(columns={'Close': '^gspccls'})
cy1.dropna()
cy1.shape
trs = pd.read_csv('https://raw.githubusercontent.com/asiqman12/ln_regression/main/w.csv')
trs.shape
for ind, row in trs.iterrows():
    trs.loc[ind, "dgsdec"] = row['DGS1MO'] / 100

import numpy as np
# pre 0.24
trs['firmcls'] = ty1['firmcl'].values
# >= 0.24
trs['firmcls'] = ty1['firmcl'].to_numpy()
trs['gspc'] = cy1['^gspccls'].values
# >= 0.24
trs['gspc'] = cy1['^gspccls'].to_numpy()
trs.dropna()

trs["gspexcess X axis"] = trs['gspc'] - trs['dgsdec']
trs["firmacess Y axis"] = trs['firmcls'] - trs['dgsdec']
trs.dropna()

trs.to_csv(r'moga.csv', index = False)
```
